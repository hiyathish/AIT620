```python
def transpose_and_continue(sequence, semitones=7, gap_beats=1.0):
    new_seq = music_pb2.NoteSequence()
    new_seq.CopyFrom(sequence)

    last_end = max(n.end_time for n in sequence.notes)
    qpm = sequence.tempos[0].qpm
    sec_per_beat = 60.0 / qpm
    gap = gap_beats * sec_per_beat

    for n in sequence.notes:
        nn = new_seq.notes.add()
        nn.pitch = n.pitch + semitones
        nn.start_time = n.start_time + last_end + gap
        nn.end_time = n.end_time + last_end + gap
        nn.velocity = n.velocity

    new_seq.total_time = max(n.end_time for n in new_seq.notes)
    return new_seq

continued = transpose_and_continue(twinkle_twinkle)
```

```python
note_seq.plot_sequence(continued)

note_seq.play_sequence(
    continued,
    synth=note_seq.midi_synth.synthesize
)
```
