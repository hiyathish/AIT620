```python
MODEL_NAME = "124M"      # GPT-2 model size
MODEL_DIR = "models"     # base directory; gpt_2_simple will create models/124M

os.makedirs(MODEL_DIR, exist_ok=True)

# Download only if not already present
if not os.path.isdir(os.path.join(MODEL_DIR, MODEL_NAME)):
    gpt2.download_gpt2(model_name=MODEL_NAME, model_dir=MODEL_DIR)
```

```python
DATASET = "pride_and_prejudice.txt"   # exact filename, no spaces

print("Exists:", os.path.isfile(DATASET))
print("Size:", os.path.getsize(DATASET))

with open(DATASET, "r", encoding="utf-8") as f:
    text = f.read()
    print("Characters:", len(text))
    print("First 200 chars:\n", text[:200])
```
